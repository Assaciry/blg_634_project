```python
from SuperPointPretrainedNetwork.demo_superpoint import SuperPointFrontend

import os, cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
superpoint = SuperPointFrontend(weights_path="./SuperPointPretrainedNetwork/superpoint_v1.pth", nms_dist=4, conf_thresh=0.015, nn_thresh=0.7)
```

```python
image = cv2.imread("./frame_00001.png", cv2.IMREAD_GRAYSCALE)
image_tensor = np.array(image, dtype=np.float32) / 255.
```

```python
pts, desc, heatmap = superpoint.run(image_tensor)
```

```python
plt.figure()
plt.imshow(image, cmap="gray")
plt.plot(pts[0,:],pts[1,:], ls="", marker=".",ms=0.8)
plt.show()
```
